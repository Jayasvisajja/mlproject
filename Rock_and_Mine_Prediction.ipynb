# **Rock and Mine Prediction Using Logistic Regression**

Rock vs Mine.png

### Overview:
Rock and mine prediction is a classification problem commonly tackled using logistic regression, a statistical method for binary classification. The objective is to determine whether an object in sonar or seismic data is a rock (non-mineral) or a mine (metallic/mineral deposit). This is crucial in fields such as geology, mineral exploration, and defense applications.

Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and  Data Processing

```python
sonar_data = pd.read_csv('/content/Copy of sonar data.csv', header = None)
```

```python
sonar_data.head()
```

```python
# number of rows and columns
sonar_data.shape
```

```python
sonar_data.describe()  #describe -> statistical measeures of the data
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
#separating data and Lables
X = sonar_data.drop(columns = 60,axis = 1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

## **Training and Test data**

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1,stratify = Y,random_state = 1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

## **Model Training -> Logistic Regression**

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train,Y_train)
```

## **Model Evaluation**

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data :',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on test data: ",test_data_accuracy)
```
